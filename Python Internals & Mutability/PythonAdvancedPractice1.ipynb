2Ô∏è‚É£ Debug This

```python
def f(x, y=[]):
    y.append(x)
    return y
print(f(2))
print(f(3))
print(f(4))
```

‚ùå Buggy Code
def f(x, y=[]):
    y.append(x)
    return y

‚ùå Problem

y is a mutable default argument

It is created once and reused across calls

print(f(1))  # [1]
print(f(2))  # [1, 2]  üò±

‚úÖ Correct Fix (INTERVIEW-SAFE)
def f(x, y=None):
    if y is None:
        y = []
    y.append(x)
    return y

‚úÖ Correct Behavior
print(f(1))  # [1]
print(f(2))  # [2]
